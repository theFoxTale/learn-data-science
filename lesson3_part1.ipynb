```python
import pandas as pd
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
export_2015 = pd.read_excel("Russia-2015.xls", sheet_name="Partner")
```

```python
export_2015.head()
```

```python
export_2016 = pd.read_excel("Russia-2016.xlsx", sheet_name="Partner")
```

```python
export_2016.head()
```

```python
full_export = pd.concat([export_2015, export_2016])
```

```python
full_export.sample(3)
```

```python
exclude = ['Europe & Central Asia', 'East Asia & Pacific', ' World', 'Unspecified', 'Middle East & North Africa']
```

```python
filtered_export = full_export[-full_export['Partner Name'].isin(exclude)]
```

```python
filtered_export.pivot_table(index=['Year'], values="Export (US$ Thousand)", aggfunc="sum")
```

```python
pivot_export = filtered_export.pivot_table(index=['Year', 'Partner Name'], values="Export (US$ Thousand)", aggfunc="sum")\
    .sort_values(['Year', 'Export (US$ Thousand)'], ascending=[True, False])
```

```python
pivot_export.index
```

```python
pivot_export[pivot_export.index.get_level_values('Year') == 2015]
```

```python
pivot_export[pivot_export.index.get_level_values('Year') == 2016]
```
